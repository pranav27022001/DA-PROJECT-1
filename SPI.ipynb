```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as exp
import plotly.graph_objects as go
import zipfile
```

```python
zip = zipfile.ZipFile('archive.zip')
```

```python
df = pd.read_csv(zip.open('spi.csv'))
```

```python
fig = exp.bar(df, x='country', y='spi_score')
fig.update_layout(title='SPI Score for every country', xaxis_title='Country', yaxis_title='SPI_Score', height=400, width=1000, showlegend=True)
fig.show()
```

Norway has the highest SPI score with the score of 92.63 followed by Iceland, Swedan and Germany with score of 92.26, 92.15, 91.78 respectively

```python
df.head()
```

```python
legend_groups=['Basic Human Needs', 'Basic Nutrition and Medical Care']
```

```python
fig = go.Figure(layout=go.Layout(height=600, width=800, title=go.layout.Title(text="Basic Needs of Every Country")))
fig.add_trace(go.Scatter(x=df['country'], y=df['basic_human_needs'], hovertext=['country', 'basic_human_needs'], name=legend_groups[0]))
fig.add_trace(go.Scatter(x=df['country'], y=df['basic_nutri_med_care'], hovertext=['country', 'basic_nutri_med_care'], name=legend_groups[1]))
fig.update_layout(xaxis_title='Country', yaxis_title='Daily Needs')
fig.show()
```

Above Analysis shows the Basic Needs available for People of Every country. Norway is Leading again with the Top Score

```python
fig = exp.scatter(df, x="country", y="wellbeing", color="country",
                 title="Quality of Life in every Country")

fig.add_trace(
    go.Scatter(
        x=df['country'],
        y=df['opportunity'],
        hovertext=[df['country'], df['opportunity']],
        showlegend=True)
)
fig.update()
fig.show()
```

Above Analysis id done for Each and every countries. Lets look into Top countries with highest SPI Score.

```python
water = df[df['water_sanitation']>90]
water.sort_values(by='water_sanitation', ascending=False)
```

```python
fig = exp.scatter(df.query("spi_score >= 80"), 
                 x="water_sanitation", 
                 y="spi_score",
                 size="spi_score", 
                 color="country",
                 hover_name="country",
                 title= "Countries with Better Water Sanitation"
                )
fig.show()
```

Above Analysis shows the Water Sanitation of Top Country having SPI score of above 90. Top 5 countries with better water sanitation are slovenia with score 99.27, Finland with score 99.26, Croatia with score 99.06, Switzerland and Belarus with score 98.96 

```python
plt.Figure(figsize=(16, 7))
fig = exp.bar(df.query('spi_score>80'), x='country', y='personal_safety',title='Safety Rate in Country having SPI score above 80', hover_data=['country', 'personal_safety'])
fig.update_layout(xaxis_title='country', yaxis_title='Personal Safety', )
fig.show()
```

```python
inclusivness = df[['country', 'inclusiveness', 'spi_rank']].sort_values(by='inclusiveness', ascending=False)
inclusivness
```

```python
from plotly.offline import iplot
```

```python
values = dict(type = 'choropleth', 
           locations = df['country'],
           locationmode = 'country names',
           colorscale='Blues',
           z = df['spi_score'], 
           text = df['country'],
           colorbar = {'title':'Social Progress Index'})

layout = dict(title = 'Social Progress Index', 
              geo = dict(showframe = True, 
                         projection = {'type': 'azimuthal equal area'}))

figure = go.Figure(data = [values], layout=layout)
iplot(figure)
```
